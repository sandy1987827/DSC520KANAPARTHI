```python
import numpy as np
import pandas as pd
import nsfg
import first
import thinkstats2
import thinkplot

def BiasPmf(pmf, label):
    new_pmf = pmf.Copy(label=label)
    
    for x, p in pmf.Items():
        new_pmf.Mult(x,x)
    new_pmf.Normalize()
    return new_pmf

resp = nsfg.ReadFemResp()
pmf = thinkstats2.Pmf(resp.numkdhh, label='numkdhh')
thinkplot.Pmf(pmf)
thinkplot.Config(xlabel='Number of children', ylabel='PMF')
```

```python
biased_pmf = BiasPmf(pmf, label='Biased')
thinkplot.PrePlot(2)
thinkplot.Pmfs([pmf, biased_pmf])
thinkplot.Show(xlabel='Number of children', ylabel='PMF')
```

```python
print('PMF Mean:', pmf.Mean())
print('PMF Variance:', pmf.Var())

print('Biased Mean:', biased_pmf.Mean())
print('Biased Variance:', biased_pmf.Var())
```

```python
preg = nsfg.ReadFemPreg()
total_live = preg[preg.outcome == 1]
firsts = total_live[total_live.birthord == 1]
others = total_live[total_live.birthord != 1]

first_cdf = thinkstats2.Cdf(firsts.totalwgt_lb, label='first')
others_cdf = thinkstats2.Cdf(others.totalwgt_lb, label='others')

thinkplot.PrePlot(2)
thinkplot.Cdfs([first_cdf,others_cdf])
thinkplot.Show(xlabel='weight (pounds)',ylabel='CDF')

print('First Born Percentile Rank: ',first_cdf.PercentileRank(7.2))
print('Others Percentile Rank: ',others_cdf.PercentileRank(7.32))
```

```python
data = np.random.random(1000)
random_pmf = thinkstats2.Pmf(data)
thinkplot.Pmf(random_pmf, linewidth=0.1)
thinkplot.Config(xlabel='Random Number', ylabel='PMF')
```

```python
random_cdf = thinkstats2.Cdf(data)
thinkplot.Cdf(random_cdf)
thinkplot.Show(xlabel='Random Data',ylabel='CDF')
```
